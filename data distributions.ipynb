```python
leaf_file='/home/ar/Projects/nasiri/all_data/data.leaf.txt'
```

```python
import pandas as pd

leaf_all = pd.read_csv(leaf_file,sep='\t')

leaf_col_names = leaf_all.columns
leaf_names = leaf_all['name']
leaf_y = leaf_all['exp']
leaf_x = leaf_all[leaf_col_names[1:-1]]
```

```python
import numpy as np

import matplotlib.pyplot as plt

n, bins, patches = plt.hist(x=np.log(1+leaf_y), bins='auto', color='#0504aa',alpha=0.7, rwidth=0.85)
```

```python
for name in leaf_x.columns:
    plt.hist(x=np.log(1+leaf_x[name]), bins='auto', color='#0504aa',alpha=0.7, rwidth=0.85)
    plt.title(name)
    plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(32, 32))
sns.heatmap(leaf_all.corr(method='spearman'))
```

```python
fnames = set([])
for n in leaf_x.columns:
    fnames.add(n.split('_')[0])
fnames
```

```python
suf = set([])
for n in leaf_x.columns:
    suf.add('_'.join(n.split('_')[1:]))
suf
```

```python
f2name = {}
for f in fnames:
    f2name[f] = []
    for n in leaf_x.columns:
        if f in n:
            f2name[f].append(n)
    f2name[f].append('exp')
```

```python
f2name['DH']
```

```python
for f in f2name:
    temp_data = leaf_all[f2name[f]]
    plt.figure(figsize=(32, 32))
    sns.heatmap(temp_data.corr(method='spearman'))
    plt.title(f)
    plt.show()
```
